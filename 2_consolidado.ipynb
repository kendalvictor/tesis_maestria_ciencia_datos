```python
import pandas as pd
```

```python
import glob
dicc_data = {
    'dia': [],
    'titulo': [],
    'contenido': [],
    'tamanio': []
}

for file in glob.iglob('news/**/*.txt', recursive=True):
    print(file.split('\\'))
    listado = file.split('\\')
    with open(file, encoding='UTF-8') as f:
        new = f.read()
        dicc_data['dia'].append(listado[1])
        dicc_data['titulo'].append(listado[-1].split('.')[0])
        dicc_data['contenido'].append(new)
        dicc_data['tamanio'].append(len(new))
        
        
data = pd.DataFrame(dicc_data)
del dicc_data
data.head(50)
```

```python
tamnio = data.groupby(by=['dia']).agg({'tamanio': ['mean', 'min', 'max']})
tamnio
```

```python
tamnio.plot()
```

```python
contenido = data.groupby(by=['dia']).agg({'contenido': lambda _: '\n'.join(list(_))})
contenido
```

```python
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords
from PIL import Image 
import matplotlib.pylab as plt


for row in contenido.iloc[-3:, :].iterrows():
    print(row[1], type(row[1]))
    comment_words = ''
    tokens = row[1]['contenido'].split()
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        
    comment_words += " ".join(tokens)+" "
    
    wordcloud = WordCloud(
        width = 800, height = 800,
        background_color ='white',
        stopwords=stopwords.words('spanish'),
        min_font_size = 10).generate(comment_words)
    
    plt.figure(figsize = (6, 6), facecolor = None)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.tight_layout(pad = 0)
```
